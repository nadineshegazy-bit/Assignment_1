## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane is Structured Content Archetype Clustering.

This is an unsupervised machine learning task because there are no predefined labels. The goal is to group content pages with similar performance characteristics into meaningful clusters that can support content management decisions.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

There is no target label because clustering is an unsupervised learning task.

The model will use observed features such as impressions, clicks, CTR, sessions, average position, content age, and engagement metrics to discover natural groups of similar pages.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The clustering is successful if the clusters are easy to interpret and each cluster represents a meaningful content archetype. A good result is when each cluster can be linked to a clear business action such as Protect, Improve, Refresh, or Monitor.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

One row represents one page.


## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Content performance depends on many features that interact with each other. A simple rule cannot capture these complex relationships. Clustering can automatically discover patterns and group similar pages without requiring manually defined rules.
